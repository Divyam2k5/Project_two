```python
fact_trips = spark.table("gold_fact_trips")
dim_riders = spark.table("gold_dim_riders")
dim_stations = spark.table("gold_dim_stations")
dim_payments = spark.table("gold_dim_payments")
```

```python
fact_trips.count()
```

```python
dim_riders.groupBy("is_member").count().show()
```

```python
from pyspark.sql.functions import col, sum

dim_payments = dim_payments.withColumn(
    "amount",
    col("amount").cast("double")
)

dim_payments.groupBy("rider_id") \
.agg(sum("amount").alias("total_amount")) \
.show()
```
