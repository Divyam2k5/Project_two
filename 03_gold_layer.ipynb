```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
trip_df = spark.table("silver_trips")
rider_df = spark.table("silver_riders")
station_df = spark.table("silver_stations")
payment_df = spark.table("silver_payments")
```

```python
# DIM RIDERS

dim_riders = rider_df.select(
    "rider_id",
    "first_name",
    "last_name",
    "address",
    "birthday",
    "account_start_date",
    "is_member"
)

# DIM STATIONS

dim_stations = station_df.select(
    "station_id",
    "station_name",
    "latitude",
    "longitude"
)

# DIM PAYMENTS

dim_payments = payment_df.select(
    "payment_id",
    "payment_date",
    "amount",
    "rider_id"
)
```

```python
fact_trips = trip_df.select(
    "ride_id",
    "rideable_type",
    "started_at",
    "ended_at",
    "start_station_id",
    "end_station_id",
    "rider_id"
)
```

```python
dim_riders.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_dim_riders")

dim_stations.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_dim_stations")

dim_payments.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_dim_payments")

fact_trips.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_fact_trips")
```
