```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
trip_df = spark.table("default.trips")

rider_df = spark.table("default.riders")

station_df = spark.table("default.stations")

payment_df = spark.table("default.payments")
```

```python
trip_df = trip_df.toDF(
    "trip_id",
    "rideable_type",
    "started_at",
    "ended_at",
    "start_station_id",
    "end_station_id",
    "rider_id"
)
```

```python
trip_df.show(5)

rider_df.show(5)

station_df.show(5)

payment_df.show(5)
```

```python
# FIX COLUMN NAMES

trip_df = trip_df.toDF(
    "ride_id",
    "rideable_type",
    "started_at",
    "ended_at",
    "start_station_id",
    "end_station_id",
    "rider_id"
)

rider_df = rider_df.toDF(
    "rider_id",
    "first_name",
    "last_name",
    "address",
    "birthday",
    "account_start_date",
    "is_member"
)

station_df = station_df.toDF(
    "station_id",
    "station_name",
    "latitude",
    "longitude"
)

payment_df = payment_df.toDF(
    "payment_id",
    "payment_date",
    "amount",
    "rider_id"
)

# CREATE BRONZE TABLES

trip_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_trips")

rider_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_riders")

station_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_stations")

payment_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_payments")
```

```python
# DROP OLD BRONZE TABLES

spark.sql("DROP TABLE IF EXISTS bronze_trips")
spark.sql("DROP TABLE IF EXISTS bronze_riders")
spark.sql("DROP TABLE IF EXISTS bronze_stations")
spark.sql("DROP TABLE IF EXISTS bronze_payments")
```
