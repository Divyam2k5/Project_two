```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
trip_df = spark.table("bronze_trips")
rider_df = spark.table("bronze_riders")
station_df = spark.table("bronze_stations")
payment_df = spark.table("bronze_payments")
```

```python
# REMOVE NULLS

trip_df = trip_df.dropna()
rider_df = rider_df.dropna()
station_df = station_df.dropna()
payment_df = payment_df.dropna()

# REMOVE DUPLICATES

trip_df = trip_df.dropDuplicates()
rider_df = rider_df.dropDuplicates()
station_df = station_df.dropDuplicates()
payment_df = payment_df.dropDuplicates()
```

```python
trip_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("silver_trips")

rider_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("silver_riders")

station_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("silver_stations")

payment_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("silver_payments")
```
